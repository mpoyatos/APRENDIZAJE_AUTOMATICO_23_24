```python
import plotly.express as px

df = px.data.gapminder()
df
```

```python
gapminder = px.data.gapminder()
gapminder2007 = gapminder.query("year == 2007")
gapminder2007.head()
```

```python
fig = px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", hover_name="country")
fig.show()
```

```python
fig = px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", color="continent",
                 hover_name='country', labels={"gdpPercap":"GDP per Capita (USD)",
                                              "lifeExp":"Life Expectation"},
                 title="Life Expectancy vs GDP per Capita",
                 size="pop",size_max=60)
fig.show()
```

```python
fig = px.scatter(gapminder2007, x="gdpPercap", y="lifeExp", color="continent",
                 hover_name='country', labels={"gdpPercap":"GDP per Capita (USD)",
                                              "lifeExp":"Life Expectation"},
                 title="Life Expectancy vs GDP per Capita",
                 size="pop",size_max=60, facet_col="continent")
fig.show()
```

```python
fig2 = px.choropleth(df, locations="iso_alpha", color="lifeExp",
                 hover_name='country', hover_data=['lifeExp', 'gdpPercap'],
                 title="TITULO",
                 color_continuous_scale=px.colors.sequential.Plasma,
                 projection="natural earth", animation_frame="year")
fig2.show()
```

```python
import altair as alt

# Calculate the mean life expectancy and gdpPercap for each continent
continent_means = df.groupby('continent')[['lifeExp', 'gdpPercap']].mean().reset_index()

# Create a scatter plot of life expectancy vs. GDP per capita, colored by continent
# and with smaller point size representing population
chart = alt.Chart(df).mark_point(size=5).encode(  # Reduced point size here
    x='gdpPercap',
    y='lifeExp',
    color='continent',
    tooltip=['country', 'lifeExp', 'gdpPercap', 'pop']
)

# Create a scatter plot of the continent means
mean_chart = alt.Chart(continent_means).mark_point(size=100, shape='diamond', color='black').encode(
    x='gdpPercap',  # Now using the calculated mean gdpPercap
    y='lifeExp',
    tooltip=['continent', 'lifeExp', 'gdpPercap']  # Added gdpPercap to tooltip
)

# Combine the two charts
(chart + mean_chart).properties(
    width=600,
    height=400
)
```

```python
import altair as alt

alt.Chart(df).mark_line().encode(
    x='year',
    y='mean(lifeExp)',
    color='continent'
).properties(
    title='Life Expectancy Over Time by Continent'
)
```

```python
alt.Chart(df).mark_boxplot().encode(
    x='gdpPercap',
    y='continent'
).properties(
    title='GDP per Capita Distribution by Continent'
)
```

```python
alt.Chart(df).mark_area().encode(
    x='year',
    y='sum(pop)',
    color='continent'
).properties(
    title='Population Growth Over Time'
)
```

```python
import altair as alt

selection = alt.selection_single(fields=['year'], bind='legend', init={'year': 1952})

alt.Chart(df).mark_point().encode(
    x='gdpPercap',
    y='lifeExp',
    color='continent',
    size='pop',
    tooltip=['country', 'lifeExp', 'gdpPercap', 'pop']
).add_selection(
    selection
).transform_filter(
    selection
).properties(
    title='Interactive Chart with Year Selection'
)
```
